## Rilevazione con algoritmi di intelligenza artificiale






Il progetto Seminiamo si avvale della collaborazione con l'istituto di ricerca Dyrecta Lab per monitorare l'impatto complessivo della riqualificazione dell'area verde nel tessuto urbano.

Un dato centrale per valutare l'efficacia dell'intervento è legato alla frequentazione dell'area verde. In assenza di supporti tecnologici, le rilevazioni venivano effettuate di persona, recandosi nella piazza una volta al giorno per contare il numero di persone presenti. Una metedologia corretta, ma dispendiosa in termini di risorse umane e poco efficace per valutare l'effettiva frequentazione della piazza durante il giorno.

Ogni piazza è un luogo vivo, aperto ai cambiamenti, e soggetto a dinamiche complesse che possono variare anche di ora in ora.

L' **algoritmo di rilevazione** implementato nel sistema, basato sulle reti neurali più efficienti nello stato dell'arte, permette un ampio margine di miglioramento rispetto all'osservazione diretta.

L’inquinamento dell’aria è emerso negli ultimi anni come una delle maggiori problematiche per la salute di chi vive in città, soprattutto in quanto è nelle aree fortemente urbanizzate che si viene a concentrare la maggior presenza di traffico automobilistico e veicolare in genere.

Il termine particolato (particulate matter, PM) individua l’insieme dei particelle con l’esclusione dell’acqua, disperse in atmosfera.

A differenza di altre forme di esposizione a rischi ambientali, l’esposizione agli inquinanti presenti nell’aria non può essere facilmente evitata. I livelli di inquinamento, tuttavia, possono presentare marcate differenze locali, specialmente per quanto riguarda le emissioni “a bassa quota” (come ad esempio quelle da trasporto stradale),  a causa di variazioni nella velocità e direzione del vento, della presenza di precipitazioni atmosferiche (pioggia o neve) o anche a causa della morfologia del territorio.

* PM10 è la frazione di particelle con diametro uguale o inferiore a 10 µm. 

* PM2.5 è la frazione di particelle con diametro uguale o inferiore a 2.5 µm. 

I limiti di legge per la protezione della salute umana sono (cfr [Arpa Lombardia](https://www.arpalombardia.it/Pages/Aria/Inquinanti/PM10-PM2,5.aspx?firstlevel=Inquinanti)):

* PM2.5 Limite annuale: 25 µg/m³ media annua 

* PM10 Limite Annuale: 40 µg/m³ media annua


Attualmente la media di pm10 è leggermente superiore rispetto ai limiti di legge (linee rosse nel grafico). Un confronto con i dati medi di Milano dello stesso periodo (fonte:Arpa) connota l'aria come particolarmente soggetta a questo tipo di inquinamento, dovuto soprattuto al traffico veicolare.
